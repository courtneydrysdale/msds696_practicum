## Separating Lyrics Files

In this notebook, I created separate dataframes for the following subsections of the dataset, to make it easier to manage the large file sizes for processing in future steps.

- English lyric songs
- Pop songs
- Rock songs

```python
import pandas as pd
```

```python
df = pd.read_csv("song_lyrics.csv")
```

```python
df_en = df[df['language'] == 'en']
```

```python
df_pop = df_en[df_en['tag'] == 'pop']
```

```python
df_rock = df_en[df_en['tag'] == 'rock']
```

```python
df_pop.reset_index(drop=True, inplace=True)
```

```python
df_rock.reset_index(drop=True, inplace=True)
```

```python
df_en.to_csv('english_lyrics.csv')
```

```python
df_pop.to_csv('pop_lyrics.csv')
```

```python
df_rock.to_csv('rock_lyrics.csv')
```
